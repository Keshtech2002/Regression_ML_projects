```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv("crim.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for n in df.columns:
    if df[n].isnull().sum() > 0:
        df[n].fillna(df[n].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['CRIM','ZN','TAX','B']])
```

```python
new = pd.DataFrame(scaled, columns=['crim','zn','tax','b'])
```

```python
df.head()
```

```python
for s in new.columns:
    df[s] = new[s]
```
