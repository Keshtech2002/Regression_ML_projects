# **IMPORTING LIBRARIES**

```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# **READING THE DATASET**

```python
#loading the dataset

from sklearn.datasets import load_boston
df = load_boston()
```

```python
#displaying the dataset

df 
```

> The dataset is in the  form of key,value pairs - like a dictionary

```python
#extracting only the independent variables (FEATURES)

pd.DataFrame(df.data)
```

```python
#storing the independent variables into a dataframe

dataset = pd.DataFrame(df.data)
```

```python
dataset.columns = df.feature_names
```

```python
#displaying the final dataset with column names

dataset
```

```python
#displaying only the first 5 records of the dataset

dataset.head()
```

> Let us now create a column ' price' for the independent feature (LABEL)

```python
#creating label (dependent variable) -- price 

dataset['price'] = df.target
```

# **DISPLAYING THE DATASET**

```python
#displaying the first 5 records of the dataset

dataset.head()
```

# **SPLITTING DATASET INTO FEATURES AND LABELS**

> **Splitting the dataset into dependent features (LABEL) - 'price' and independent features (all columns except 'price')**

```python
#splitting the dataset into dependent features (LABEL) - 'price' and independent features (all columns except 'price')

x = dataset.drop(columns=['price'])
y = dataset['price']
```

```python
#displaying the dependent feature

y
```

```python
#displaying the independent feature

x
```

# **LINEAR REGRESSION**

```python
#Importing Linear Regression and Cross Validation from sklearn library

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

**MODEL CREATION**

```python
#model 

lr = LinearRegression()
```

**CROSS VALIDATION**

> Cross Validation takes different combination of the dataset to train the model , let us take 5 as cv for our model

```python
#Performing Cross Validation on the dataset

mse = cross_val_score(lr,x,y,scoring='neg_mean_squared_error',cv=5)
```

```python
#5 cross validations performed

mse
```

```python
#calculating mean of the mse

mean_mse = np.mean(mse)
mean_mse
```

> **We got -37 score for the Linear Regression model that we have created**

# **RIDGE REGRESSION**

```python
#Importing Ridge Regression from sklearn library and GridSearchCV for Hyper Parameter Tuning

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV  #hyperparameter tuning
```

**MODEL CREATION**

```python
#model

ridge = Ridge()
```

```python
#creating parameter for Hyper Paramater Tuning

params={'alpha' :[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
```

```python
#object creation

ridge_regressor = GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=10)
```

```python
#fitting our model 

ridge_regressor.fit(x,y)
```

**BEST PARAMETER AND SCORE OF THE MODEL**

```python
ridge_regressor.best_params_
```

```python
ridge_regressor.best_score_
```

> **We got -29 score for the Ridge Regression model that we have created**
