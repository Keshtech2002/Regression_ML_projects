# **EDA**

```python
#heatmap to visualize missing values

sns.heatmap(df.isnull())
```

```python
#correlation analysis

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot = True, cmap='coolwarm')
```

```python
df
```

```python
#checking fot skewness of the dataset

skew = df.skew().sort_values(ascending = False)
skew
```

If the skewness value is between 0.5 and -0.5 then it is normal distribution else will be right of left depending upon data.

```python
#BOXPLOT - checking for outliers

plt.figure(figsize=(10,10))
sns.boxplot(data=df)
plt.show()
```

For a better understanding let us draw a boxplot for each feature in the dataset 

```python
# individual box plot

def Box(df):
    plt.title("CRIM Box Plot")
    sns.boxplot(df)
    plt.show()
Box(df['CRIM'])
```

This is a right skewed data as the outliers are depicted on the right hand side of the boxplot
